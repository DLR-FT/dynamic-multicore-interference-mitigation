```python
%matplotlib ipympl
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse_dt_df(string):
    # Define the regular expression pattern to match the dt/df ratio
    pattern = r'(\d+)/(\d+)'

    # Use the re.search() function to find the match
    match = re.search(pattern, string)

    if match:
        dt = int(match.group(1))
        df = int(match.group(2))
        return {"dt": dt, "df": df}
    else:
        return None
```

```python
with open("foo1.txt") as f:
    lines = f.readlines()

data = []
for line in lines:
    data.append(parse_dt_df(line))

plt.figure()
plt.plot( [x["dt"] for x in data])
```
